```python
dbutils.widgets.text("catalog","","Enter Catalog")
CATALOG=dbutils.widgets.get("catalog")
dbutils.widgets.text("schema","","Enter Schema")
SCHEMA=dbutils.widgets.get("schema")
```

```python
spark.sql(f"CREATE CATALOG IF NOT EXISTS {CATALOG}")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {CATALOG}.{SCHEMA}")
print(f"{CATALOG}.{SCHEMA}")
spark.sql(f"CREATE VOLUME IF NOT EXISTS {CATALOG}.{SCHEMA}.datalake")
spark.sql(f"CREATE VOLUME IF NOT EXISTS {CATALOG}.{SCHEMA}.bronze")
spark.sql(f"CREATE VOLUME IF NOT EXISTS {CATALOG}.{SCHEMA}.silver")
spark.sql(f"CREATE VOLUME IF NOT EXISTS {CATALOG}.{SCHEMA}.gold")
```

```python
SRC = f"/volume/{CATALOG}/{SCHEMA}/datalake"
BRONZE = f"/volume/{CATALOG}/{SCHEMA}/bronze"
SILVER = f"/Volumes/{CATALOG}/{SCHEMA}/silver"
GOLD = f"/Volumes/{CATALOG}/{SCHEMA}/gold"
SILVERDB=f"{CATALOG}.{SCHEMA}"
GOLDDB   = f"{CATALOG}.{SCHEMA}"
print("source datalake location",SRC)
print("bronze datalake location",BRONZE)
print("silver database",SILVERDB)
print("gold database",GOLDDB)
```

```python
#used to return multiple values from one notebook to another notebook.
#It packages several variables into a JSON object and sends them back to the calling notebook.
#dumps() It converts a Python dictionary into a JSON string and josn is used because notebook.exit() can return only a string.

import json
dbutils.notebook.exit(
    json.dumps({
        "CATALOG": CATALOG,
        "SCHEMA": SCHEMA,
        "SRC": SRC,
        "BRONZE": BRONZE,
        "SILVER": SILVER,
        "GOLD": GOLD,
        "SILVERDB": SILVERDB,
        "GOLDDB": GOLDDB
    })
)
#Child notebook sends JSON
#Parent notebook receives JSON
#Parent parses JSON into dict
```
