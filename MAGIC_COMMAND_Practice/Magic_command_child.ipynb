```python
%python
dbutils.widgets.text("Name","Unkown")
name=dbutils.widgets.get("Name")
print("The name you entered is:",name)
```

```python
dbutils.widgets.dropdown("Gender","Male",["Male","Female","GAY"])
gender=dbutils.widgets.get("Gender")
print("The gender you entered is:",gender)
```
