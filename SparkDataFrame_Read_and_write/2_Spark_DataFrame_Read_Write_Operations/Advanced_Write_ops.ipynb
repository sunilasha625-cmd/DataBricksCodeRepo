**Below example shows PERMISSIVE**  

```python
from pyspark.sql.types import *

schema_with_corrupt = StructType([
    StructField("cust_id", IntegerType(), True),
    StructField("cust_name", StringType(), True),
    StructField("age", IntegerType(), True),
    StructField("amount", IntegerType(), True),
    StructField("txn_date", DateType(), True),
    StructField("_bad_record", StringType(), True)
])
```

```python
df = (
    spark.read
    .schema(schema_with_corrupt)
    .option("header", "true")
    .option("mode", "PERMISSIVE")
    .option("inferSchema", "true")
    .option("columnNameOfCorruptRecord", "_bad_record")
    .csv("/Volumes/workspace/default/volumewd36/fast_fail.txt")
)
```

```python
df.printSchema()
```

```python
df.count()
```

```python
df.select("*").show(20,False)
```

**Below example for DROPMALFORMED**

```python
df1 = (
    spark.read
    .schema(schema_with_corrupt)
    .option("header", "true")
    .option("mode", "DROPMALFORMED")
    .option("columnNameOfCorruptRecord", "_bad_record")
    .csv("/Volumes/workspace/default/volumewd36/fast_fail.txt")
)
```

```python
df1.select("*").show(20,False)
```
