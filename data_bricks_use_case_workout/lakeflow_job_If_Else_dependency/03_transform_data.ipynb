```python
#Run only if validation PASS

from pyspark.sql.functions import sum

df = spark.read.table("sales.sales.sales_raw")

agg_df = df.groupBy("region") \
           .agg(sum("amount").alias("total_sales"))

agg_df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("sales.sales.sales_summary")
```
